```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Cargamos el dataset
df = pd.read_csv("/content/drive/MyDrive/prueba_proyecto.csv", sep=';')
# Verificamos la estructura del dataset
print("Primeras filas del dataset:")
print(df.head())
print("\nDescripción del dataset:")
print(df.describe())

print("\nInformación del dataset:")
print(df.info())
# Conversión de tipos de datos
df['ano'] = pd.to_numeric(df['ano'], errors='coerce')
df['semana'] = pd.to_numeric(df['semana'], errors='coerce')
# Reemplazamos los valores NaN con 0 para el análisis de sumas
df.fillna(0, inplace=True)
# Análisis exploratorio
print("\nNúmero de casos de IRAS no neumonías:")
print(df['ira_no_neumonia'].sum())
print("\nNúmero de episodios de neumonías en menores de 5 años:")
print(df['neumonias_men5'].sum())
print("\nNúmero de episodios de neumonías en mayores de 60 años:")
print(df['neumonias_60mas'].sum())
print("\nNúmero de hospitalizaciones por neumonías en menores de 5 años:")
print(df['hospitalizados_men5'].sum())
print("\nNúmero de hospitalizaciones por neumonías en mayores de 60 años:")
print(df['hospitalizados_60mas'].sum())
print("\nNúmero de defunciones por neumonías en menores de 5 años:")
print(df['defunciones_men5'].sum())
print("\nNúmero de defunciones por neumonías en mayores de 60 años:")
print(df['defunciones_60mas'].sum())
# Gráficos
# Evolución de casos de IRAS no neumonías por año
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='ano', y='ira_no_neumonia', estimator='sum', errorbar=None)
plt.title('Evolución de casos de IRAS no neumonías por año')
plt.xlabel('Año')
plt.ylabel('Número de Casos')
plt.grid(False)  # Quitar las cuadrículas
plt.show()
# Hospitalizaciones por neumonía en menores de 5 años vs mayores de 60 años
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='ano', y='hospitalizados_men5', estimator='sum', errorbar=None, label='Menores de 5 años')
sns.lineplot(data=df, x='ano', y='hospitalizados_60mas', estimator='sum', errorbar=None, label='Mayores de 60 años')
plt.title('Hospitalizaciones por neumonía')
plt.xlabel('Año')
plt.ylabel('Número de Hospitalizaciones')
plt.legend()
plt.grid(False)  # Quitar las cuadrículas
plt.show()
# Defunciones por neumonía en menores de 5 años vs mayores de 60 años
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='ano', y='defunciones_men5', estimator='sum', errorbar=None, label='Menores de 5 años')
sns.lineplot(data=df, x='ano', y='defunciones_60mas', estimator='sum', errorbar=None, label='Mayores de 60 años')
plt.title('Defunciones por neumonía')
plt.xlabel('Año')
plt.ylabel('Número de Defunciones')
plt.legend()
plt.grid(False)  # Quitar las cuadrículas
plt.show()
# Histograma de hospitalizaciones por departamento
# Agrupamos por departamento y sumar las hospitalizaciones
hospitalizaciones_por_departamento = df.groupby('departamento')[['hospitalizados_men5', 'hospitalizados_60mas']].sum().reset_index()
# Calculamos la suma total de hospitalizaciones
hospitalizaciones_por_departamento['total_hospitalizaciones'] = hospitalizaciones_por_departamento['hospitalizados_men5'] + hospitalizaciones_por_departamento['hospitalizados_60mas']
# Ordenamos por la cantidad total de hospitalizaciones de forma descendente
hospitalizaciones_por_departamento = hospitalizaciones_por_departamento.sort_values(by='total_hospitalizaciones', ascending=False)
# Creamos el histograma
plt.figure(figsize=(12, 6))
sns.barplot(data=hospitalizaciones_por_departamento, x='departamento', y='total_hospitalizaciones', palette='viridis')
plt.title('Total de Hospitalizaciones por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de Hospitalizaciones')
plt.xticks(rotation=45, ha='right')  # Ajustar rotación y alineación de etiquetas
plt.grid(False)  # Quitar las cuadrículas
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Cargamos el dataset (suponiendo que está en formato CSV)
df = pd.read_csv("/content/drive/MyDrive/prueba_proyecto.csv", sep=";")

# Seleccionamos las variables predictoras y la variable objetivo
X = df[['ira_no_neumonia', 'neumonias_men5', 'ano', 'semana']]  # Variables independientes
y = df['hospitalizados_men5']  # Variable dependiente

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Creamos un modelo de regresión lineal
model = LinearRegression()

# Entrenamos el modelo
model.fit(X_train, y_train)

# Hacemos predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluamos el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Coeficientes del modelo entrenado
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)

# Gráfico del modelo de regresión lineal
plt.figure(figsize=(10, 6))

# Datos observados vs predichos
plt.scatter(y_test, y_pred, color='blue', label='Datos Observados vs Predichos')

# Línea de 45 grados para comparación
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Línea de 45°')

plt.title('Modelo de Regresión Lineal: Observado vs Predicho')
plt.xlabel('Observado')
plt.ylabel('Predicho')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("/content/drive/MyDrive/prueba_proyecto.csv", sep=";")

# Seleccionamos las variables predictoras y objetivo
X = df[['ano', 'semana', 'ira_no_neumonia', 'neumonias_men5']].values
y = df['hospitalizados_men5'].values

# Escalamos las variables
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Dividimos los datos en entrenamiento y prueba
split = int(len(df) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Redimensionamos para LSTM (Long Short-Term Memory)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Construimos el modelo LSTM (Long Short-Term Memory)
model = Sequential()
model.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(1))

# Compilamos el modelo
model.compile(loss='mse', optimizer='adam')

# Entrenamos el modelo
history = model.fit(X_train, y_train, epochs=50, batch_size=72, validation_data=(X_test, y_test), verbose=2)

# Hacemos predicciones
y_pred = model.predict(X_test)

# Filtramos los valores para eliminar el outlier (punto mayor a 40)
threshold = 40
filtered_indices = y_test <= threshold
y_test_filtered = y_test[filtered_indices]
y_pred_filtered = y_pred[filtered_indices]

# Calculamos métricas de evaluación con datos filtrados
mse_filtered = mean_squared_error(y_test_filtered, y_pred_filtered)
r2_filtered = r2_score(y_test_filtered, y_pred_filtered)
print(f'Mean Squared Error (filtered): {mse_filtered}')
print(f'R-squared (filtered): {r2_filtered}')

# Gráfico de resultados filtrados
plt.figure(figsize=(10, 6))
plt.plot(y_test_filtered, label='Datos Observados')
plt.plot(y_pred_filtered, label='Predicciones')
plt.title('Predicciones de hospitalizaciones por neumonías en menores de 5 años usando LSTM (Filtrado)')
plt.xlabel('Índice del dato')
plt.ylabel('Número de Hospitalizaciones')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

# Leemos los datos espaciales
provincias = gpd.read_file('/content/drive/MyDrive/peru_provincial.geojson')

# Leemos los datos de nuestro dataset
salud = pd.read_csv('/content/drive/MyDrive/prueba_proyecto.csv', delimiter=';')

# Normalizamos los nombres
provincias['NOMBPROV'] = provincias['NOMBPROV'].str.title()
salud['provincia'] = salud['provincia'].str.title()

# Sumamos el total de casos por provincia
salud_total = salud.groupby('provincia').agg({
    'ira_no_neumonia': 'sum'
}).reset_index()

# Renombramos las columnas en 'salud_total' para que coincidan con 'provincias'
salud_total.columns = ['NOMBPROV', 'ira_no_neumonia']

# Unimos datasets usando la columna "NOMBPROV"
df = provincias.set_index('NOMBPROV').join(salud_total.set_index('NOMBPROV'))

# Limpiamos datos: asegurar que no haya NaN ni valores negativos
df['ira_no_neumonia'].fillna(0, inplace=True)
df['ira_no_neumonia'] = df['ira_no_neumonia'].clip(lower=0)

# Definimos más rangos logarítmicos para los casos de IRAS no neumonías
rangos_log = np.geomspace(df['ira_no_neumonia'].min() + 1, df['ira_no_neumonia'].max(), num=15)  # Más rangos logarítmicos
rangos = [0] + list(rangos_log)
intervalos = ['[{:.0f}, {:.0f})'.format(rangos[i], rangos[i + 1]) for i in range(len(rangos) - 1)]
df['RANGO'] = pd.cut(df['ira_no_neumonia'], bins=rangos, labels=intervalos, right=False)

# Definimos más colores correspondientes a cada rango
colors = ['#FEEBE5', '#FCC5C0', '#FA9FB5', '#F768A1', '#DD3497', '#AE017E', '#7A0177', '#49006A', '#ABD9E9', '#74ADD1', '#2C7BB6', '#045A8D', '#8856A7', '#810F7C', '#4D004B']  # Más colores
cmap = LinearSegmentedColormap.from_list('custom_cmap', colors, N=len(colors))

# Generamos el mapa
fig, ax = plt.subplots(figsize=(12, 12))
df.plot(column='RANGO', cmap=cmap, ax=ax, edgecolor='0.8', legend=False)

# Personalizamos
ax.set_title('Mapa de calor de IRAS no neumonías por provincia')
ax.set_axis_off()

# Definimos la leyenda de colores
cbar = fig.colorbar(plt.cm.ScalarMappable(cmap=cmap), ax=ax, shrink=0.6)
cbar.set_ticks([i / (len(intervalos) - 1) for i in range(len(intervalos))])
cbar.set_ticklabels(intervalos)

# Mostramos
plt.show()


```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("/content/drive/MyDrive/prueba_proyecto.csv", sep=";")

# Seleccionamos las variables predictoras y la variable objetivo
X = df[['ano', 'semana', 'ira_no_neumonia', 'neumonias_men5']].values
y = df['hospitalizados_men5'].values

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Creamos un modelo de Árbol de Decisión
tree_model = DecisionTreeRegressor(random_state=0)

# Entrenamos el modelo
tree_model.fit(X_train, y_train)

# Hacemos predicciones sobre el conjunto de prueba
y_pred = tree_model.predict(X_test)

# Evaluamos el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Gráfico de resultados
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Datos Observados vs Predichos')

# Línea de 45 grados para comparación
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Línea de 45°')

plt.title('Modelo de Árbol de Decisión: Observado vs Predicho')
plt.xlabel('Observado')
plt.ylabel('Predicho')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```
